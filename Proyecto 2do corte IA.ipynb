> <h2> 3.4 Entrenamiento haarcascade</h2>
    
> <p style="font-size:20px;">En esta etapa usaremos la herramienta traincascade, usaremos el archivo índice con las imágenes negativas y el archivo de muestras generado anteriormente el cual nos dará como resultado un archivo cascade.xml. Podemos realizar el entrenamiento usando la siguiente línea en la consola de Windows posicionándonos en la carpeta raíz del proyecto o creando un archivo ejecutable .bat y guardándolo en la carpeta raíz del proyecto con la misma línea de comando:</p>
>![Papel y lápiz](imagen11.png "MarkDown")

>![Papel y lápiz](imagen13.jpeg "MarkDown")

> <h2>C:\opencv\build\x64\vc15\bin\opencv_traincascade.exe -data data/cascade -vec data/muestra.vec -bg negativas.txt -numPos 32 -numNeg 17 -numStages 25 -minHitRatio 0.999 -maxFalseAlarmRate 0.5 -precalcValBufSize 7000 -precalcIdxBufSize 7000 -w 24 -h 24 -mode ALL </h2>
 
 
><ul>
>  <li><p style="font-size:20px;">data: directorio de salida para la generación del entrenamiento. </p></li>
>  <li><p style="font-size:20px;">vec: archivo con muestras generado en la etapa anterior. </p></li>
>  <li><p style="font-size:20px;">bg: archivo índice con las imágenes negativas.</p></li>
>  <li><p style="font-size:20px;">numPos: numero de imagenes positivas indexadas. </p></li>
>  <li><p style="font-size:20px;">numNeg: numero de imagenes negativas indexadas. </p></li>
>  <li><p style="font-size:20px;">numStages: número de etapas de entrenamiento. Es directamente 	proporcional a la calidad del clasificador a generar. </p></li>
>  <li><p style="font-size:20px;">minHitRatio: minimo rando de cada etapa. </p></li>
>  <li><p style="font-size:20px;">maxFalseAlarmRate: determina el rango maximo de falsas alarmas admitidas para cada etapa.</p></li>
>  <li><p style="font-size:20px;">precalcValBufSize: tamaño del almacenamiento intermedio para valores de características precalculadas (en Mb).</p></li>
>  <li><p style="font-size:20px;">precalcIdxBufSize: tamaño del búfer para índices de características precalculadas (en Mb). Cuanta más memoria tenga, más rápido será el proceso de entrenamiento.</p></li>
>  <li><p style="font-size:20px;">w: ancho de la muestra de salida.</p></li>
>  <li><p style="font-size:20px;">h: alto de la muestra de salida. </p></li>
></ul>

    En la ruta proyecto/data/cascade podremos encontrar nuestro archivo cascade.xml el cual podremos usar para el reconocimiento.

```python
# Ejemplo de deteccion facial con OpenCV y Python
# Por Glare
# www.robologs.net

import numpy as np
import cv2
import sys

# cargamos la plantilla e inicializamos la webcam:
face_cascade = cv2.CascadeClassifier('cascade2.xml')
cap = cv2.VideoCapture(0)

while (True):
    # leemos un frame y lo guardamos
    ret, img = cap.read()

    # convertimos la imagen a blanco y negro
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # buscamos las coordenadas de los rostros (si los hay) y
    # guardamos su posicion
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Dibujamos un rectangulo en las coordenadas de cada rostro
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Mostramos la imagen
    cv2.imshow('img', img)

    # con la tecla 'q' salimos del programa
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
